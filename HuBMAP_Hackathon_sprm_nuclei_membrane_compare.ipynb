```r
# linked datasets
uuids = 'ff77fcae7f6d9b5b7b8741c282677eef'
```

```r
dir <- paste0('datasets/', uuids, '/sprm_outputs/')
```

```r
pos_file <- paste0(dir, 'reg001_expr.ome.tiff-cell_centers.csv')
pos <- read.csv(pos_file, row.names=1)
pos <- pos[,1:2]
head(pos)
```

```r
# read in the nuclear protein expression
nuc_file <- paste0(dir, 'reg001_expr.ome.tiff-nuclei_channel_mean.csv')
nuc_mat <- read.csv(nuc_file, row.names=1)
head(nuc_mat)
```

```r
membrane_file <- paste0(dir, 'reg001_expr.ome.tiff-cell_channel_meanAll.csv')
membrane_mat <- read.csv(membrane_file, row.names=1)
head(membrane_mat)
```

```r
print(dim(pos))
print(dim(nuc_mat))
print(dim(membrane_mat))
```

```r
cells_have <- intersect(rownames(pos), rownames(nuc_mat))
length(cells_have)
```

```r
library(ggplot2)
df <- data.frame(pos[cells_have,], protein = nuc_mat[cells_have, 'Hoechst1'])
ggplot(df, aes(x=x, y=y, col=protein)) + geom_point() + scale_color_viridis_c() + ggtitle('Hoechst1 in nucleus')

df <- data.frame(pos[cells_have,], protein = membrane_mat[cells_have, 'Hoechst1'])
ggplot(df, aes(x=x, y=y, col=protein)) + geom_point() + scale_color_viridis_c() + ggtitle('Hoechst1 in membrane')
```

```r
df <- data.frame(nuc = colMeans(nuc_mat[cells_have,]), membrane=colMeans(membrane_mat[cells_have, ]), label=colnames(nuc_mat))
ggplot(df, aes(x=nuc, y=membrane, label=label)) + geom_point() + geom_text(aes(label = label))
```
